#**1.- Comprension del Negocio**

Un gerente de banco se siente incómodo con que cada vez más clientes abandonen sus servicios de tarjeta de crédito. Realmente agradecerían que alguien pudiera predecir quién se verá afectado para poder acudir de manera proactiva al cliente para brindarle mejores servicios y cambiar las decisiones del cliente en la dirección opuesta.

> **Objetivo**: 
>
> Usando modelos de ML se busca predeccir el abandono de servicios de la tarjeta de credito

### **> Problema**

- En qué medida un modelo de predicción de fuga ayudará al banco en la retención de clientes.
- El consecutivo aumento de casos de desercion de servicios de tarjeta de credito en el banco

### **> Hipótesis**
- Un modelo de predicción de fuga de cliente ayudará al banco en la retención de clientes.
- Un modelo de clasificacion para saber que perfiles de usuarios que dejan la tarjeta de credito.


**Problema:** ¿Como evitariamos que los clientes del Banco 'X' cancelen su servicio de tarjetas de crédito?
1. Realizar un modelo de machine learning para el Attrition mejoraria la identificacion de clientes que cancelan su servicio de tarjeta de credito.
2. Realizar un modelo para identificar si un cliente utilizara o no su tarjeta disminuiria la tasa de clientes que cancelan su servicio de tarjetas de crédito.
3. Aumentar la linea de crédito a los clientes que tienen mas tiempo con el banco disminuiria la tasa de clientes que cancelan su servicio de tarjetas de crédito.

```python
def pltCountplot(cats,data):
    
    fig, axis = plt.subplots(len(cats) // 3,3, figsize=(20,12))  

    index = 0
    for i in range(len(cats) // 3):
        for j in range(3):
            
            ax = sns.countplot(cats[index], data=data, ax=axis[i][j])
            
            if cats[index] in ['Education_Level', 'Income_Category']:
                for item in ax.get_xticklabels():
                    item.set_rotation(15)
                
            for p in ax.patches:
                height = p.get_height()
                ax.text(p.get_x()+p.get_width()/2.,
                        height + 3,
                        '{:1.2f}%'.format(height/len(data)*100),
                        ha="center") 
            index += 1

pltCountplot(varCategoricas,banco)
```

```python
groupEtapa = banco.groupby(by=['Education_Level','Attrition_Flag']).count()
#groupEtapa
group_Etapa = groupEtapa['CLIENTNUM']
# group_Etapa
```

```python
# Convirtiendo lo anterior agrupamiento a un dataframe
df_Etapa=pd.DataFrame(group_Etapa)
df_Etapa
check=df_Etapa.reset_index(inplace=True)
df_Etapa
```

```python
intervalo_etapa=group_Etapa.index.levels[0]
```

```python
def serieGraficoBarDouble(x,intervalo):
    personasM = dict()
    personasF = dict()
    arrayNew = ["M","F"]

    for element in intervalo:
        m = x.loc[(x["Education_Level"] == element )  & (x["Attrition_Flag"] == "Attrited Customer")]['CLIENTNUM'].sum()
        f = x.loc[(x["Education_Level"] == element )  & (x["Attrition_Flag"] == "Existing Customer")]['CLIENTNUM'].sum()
        personasM[element] = m
        personasF[element] = f
        
    print("personasM",personasM)
    print("personasF",personasF)
    
    return pd.Series(personasM),pd.Series(personasF)

intervaloM,intervaloF = serieGraficoBarDouble(df_Etapa,intervalo_etapa)
```

```python
#FUGA SEGUN EL NIVEL DE EDUCACION#
(intervaloF).plot.barh(color="#E35440" , figsize=(10,8),label = 'Existing');
(intervaloM*-1).plot.barh(color="#5F92EC" , figsize=(10,8),label = 'Attrited');
   
n = 12
Y = np.arange(n)
for y,valueM,valueF in zip(Y,intervaloM*-1,intervaloF):
    pl.text(valueF +1500,y,valueF, ha='center', va='center',fontsize='large')
    pl.text(valueM -1400,y,valueM*-1, ha='center', va='center',fontsize='large')

# plt.xlim(-15000, +18000)
pl.xlim(-10000, +10000)
plt.title('GRAFICO DE LEVEL DE EDUCACION VS ESTADO DEL SERVICIO DE LA TARJETA DE CREDITO', fontdict={'fontsize':16})
plt.ylabel('LEVEL DE EDUCACION',fontsize='large')
plt.xlabel('CANTIDAD DE PERSONAS QUE CUENTAN O NO CON EL SERVICIO DE TARJETA DE CREDITO',fontsize='large')
plt.xticks([])
plt.legend(loc='upper right', bbox_to_anchor=(1, 1),
           borderpad=0.4,fontsize='x-large',
           title="Género",title_fontsize=14)
plt.show()
```
